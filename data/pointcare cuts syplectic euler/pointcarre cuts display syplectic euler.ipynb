```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
for i in os.listdir():
    if '.npy' in i:
        fname = i.split('_')
        title = fname[3]
        title = float(title)
        title = round(title,3)
        dta = np.load(i,allow_pickle=True)
        dta = dta.T
        plt.plot(dta[0],dta[1],'.k',markersize=0.8)
        plt.title('Energy = {}'.format(title))
        plt.show()
```
